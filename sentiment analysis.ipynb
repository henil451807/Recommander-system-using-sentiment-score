# Make function for classify updated_score

```python
def classify_scores(a):
    if a <=-0.25:
        v=1
    elif a <=1:
        v=2
    elif a <=2:
        v=3
    elif a <=3:
        v=4
    elif a <=5:
        v=5
    return v
```

# Apply function on updated_score and put into New_score column

```python
df['New_score'] = df['Updated_score'].apply(classify_scores)
df['New_score'] = pd.to_numeric(df['New_score'])
df
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import re
```

```python
def update_text(text):
    try:
        stop_word = set(stopwords.words('english'))
        punctuation = string.punctuation
        word_token = word_tokenize(text)
        filtered_list = [w for w in word_token if not w.lower() in 
stop_word]
        clean_words_list = [w for w in filtered_list if w not in 
punctuation]
        clear_text=" ".join(str(elem) for elem in clean_words_list)
        result = re.sub(r'\d+', '', clear_text)
        return result
    except:
        return None
```

```python
df['Updated_text'] = df['Text'].apply(update_text)
df
```

```python
from textblob import TextBlob
```

```python
def sentiments1(text):
    try:
        return TextBlob(str(text)).sentiment.polarity
    except:
        return None

df['sentiments1'] = df['Updated_text'].apply(sentiments1)
df['updated_score1'] = df['Rating']*df['sentiments1']
df
```

# Find nearest neighbours and their distance values with their indices.

# Create dictionary and store recommendations of productID which is given by USER

```python
data_dict={}
for i in range(0, len(similarity.flatten())):   

    if i == 0:
        print('Recommendations for {0}:\n'.format(df_pivot.index[query_index]))
    else:
        data_dict[str(df_pivot.index[indices.flatten()[i]])] = float(similarity.flatten()[i])
        print(f'{df_pivot.index[indices.flatten()[i]]}, is similarity distance = with {similarity.flatten()[i]:.20f}:')

print(data_dict)
```
